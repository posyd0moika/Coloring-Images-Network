```python
base_model = tf.keras.applications.ResNet50(include_top=False,
                                            input_shape=(224, 224, 3),
                                            pooling='avg', classes=100,
                                            weights='imagenet')
base_model.trainable = False

inp = Input((224, 224, 1))

cl_x = Conv2D(3, (3, 3), activation='relu', padding='same')(inp)  # 3 224 224
cl_x = base_model(cl_x)

hidden_dense_cl = Dense(512, "relu")(cl_x)

output_cl = Dense(100, "softmax")(hidden_dense_cl)

model_cl = tf.keras.Model(inp, output_cl)
model_cl.summary()
```

```python
class Inception(tf.keras.layers.Layer):
    def __init__(self, chanel: int, kernels: list[tuple], activation="relu", padding="same", strides=(1, 1)):
        super(Inception, self).__init__()

        self.layers = [
            Conv2D(chanel, kr, activation=activation, padding=padding, strides=strides)
            for kr in kernels
        ]

    def call(self, X):
        return concatenate([layer(X) for layer in self.layers])


class ConcatConvTranspose(tf.keras.layers.Layer):
    def __init__(self, X, chanel, layer, kernel=(3, 3)):
        super(ConcatConvTranspose, self).__init__()

        self.ConvTranspose = Conv2DTranspose(chanel, kernel, strides=(2, 2), padding="same")
        self.Conv = layer

        self.fl_first = True
        self.X_vec = X

    def call(self, X):
        if self.fl_first:
            _, s1, s2, s3 = X.shape
            print(X.shape)
            self.X_vec = Reshape((s1, s2, 25088 // (s1 * s2)))(self.X_vec)

            self.fl_first = False
        conc = concatenate([X, self.X_vec])
        return conc
        x = self.ConvTranspose(conc)
        return self.Conv(x)



```
